```python
import matplotlib.pyplot as plt
import pandas as pd

import torch
import torch.nn as nn
from torch.optim import Adam
```

```python
class Net(nn.Module):
    """maps from time to positive numbers"""

    def __init__(self):
        super().__init__()
        self.lin1 = nn.Linear(1, 12)
        self.lin2 = nn.Linear(12, 12)
        self.lin3 = nn.Linear(12, 12)
        self.lin4 = nn.Linear(12, 1)

        self.act = torch.relu

    def forward(self, t):
        x = self.lin1(t)
        x = self.act(x)
        x = self.lin2(x)
        x = self.act(x)
        x = self.lin3(x)
        x = self.act(x)
        x = self.lin4(x) - 10

        return torch.exp(x)
```

```python
N = 50000.

S0 = N - 100.
I0 = 100.
D0 = 0.
R0 = 0.


St = [S0]
It = [I0]
Rt = [R0]
Dt = [D0]

dt = 1

beta_time_series = 0.11 + 0.02 * torch.sin(torch.arange(200) / 5)
alpha = 0.01
delta = 0.01
gamma_time_series = 0.1 / (1 - torch.exp(-(10 + torch.arange(200)) / 50))

for i in range(200 - 1):

    beta = beta_time_series[i]
    gamma = gamma_time_series[i]

    dS = (gamma * Rt[-1]) - (beta * St[-1] * It[-1] / N)
    dI = (beta * St[-1] * It[-1] / N) - (alpha + delta) * It[-1]
    dR = (alpha * It[-1]) - (gamma * Rt[-1])
    dD = delta * It[-1]

    S = St[-1] + (dt * dS)
    I = It[-1] + (dt * dI)
    R = Rt[-1] + (dt * dR)
    D = Dt[-1] + (dt * dD)

    St.append(S)
    It.append(I)
    Rt.append(R)
    Dt.append(D)

St = torch.tensor(St)
It = torch.tensor(It)
Rt = torch.tensor(Rt)
Dt = torch.tensor(Dt)
```

```python
plt.plot(St)
plt.plot(It)
plt.plot(Rt)
plt.plot(Dt)
```

## instantiate greeks and reverse engineer the model

```python
dI = It[1:] - It[:-1]
dD = Dt[1:] - Dt[:-1]

alpha = Net()
beta = Net()
gamma = Net()
delta = Net()
```

```python
def second_derivative_loss(input_, model):

    input_.requires_grad = True
    ft = model(input_)
    
    dftdt = torch.autograd.grad(ft.sum(), input_, create_graph=True)[0]
    df2tdt2 = torch.autograd.grad(dftdt.sum(), input_, create_graph=True)[0]

    return torch.mean(df2tdt2**2)
```

```python
optim_alpha = Adam(alpha.parameters(), lr = 1e-5)
optim_beta = Adam(beta.parameters(), lr = 1e-5)
optim_gamma = Adam(gamma.parameters(), lr = 1e-5)
optim_delta = Adam(delta.parameters(), lr = 1e-5)

S_train = torch.tensor([N - It[0] for _ in range(200)])
I_train = torch.tensor([It[0] for _ in range(200)])
R_train = torch.tensor([0. for _ in range(200)])
D_train = torch.tensor([0. for _ in range(200)])


EPOCHS = 20
T = torch.arange(199).reshape(199, 1) * 1.

for i in range(EPOCHS):

    ### generate trajectories
    for t in range(199):
        t_tensor = torch.tensor([1. * t]).reshape(1, 1)

        dS_train = gamma(t_tensor) * R_train[t] - (beta(t_tensor) * I_train[t] * S_train[t] / N)
        dI_train = (beta(t_tensor) * I_train[t] * S_train[t] / N) - ((alpha(t_tensor) + delta(t_tensor)) * I_train[t])
        dD_train = delta(t_tensor) * I_train[t]
        dR_train = (alpha(t_tensor) * I_train[t]) - gamma(t_tensor) * R_train[t]

        S_train[t + 1] = S_train[t] + dS_train
        I_train[t + 1] = I_train[t] + dI_train
        D_train[t + 1] = D_train[t] + dD_train
        R_train[t + 1] = R_train[t] + dR_train

    #print(D_train)

    ### zero grad
    optim_alpha.zero_grad()
    optim_beta.zero_grad()
    optim_gamma.zero_grad()
    optim_delta.zero_grad()
    
    ### I and D should be close to their true values
    loss_I = torch.mean(It - I_train)
    loss_D = torch.mean(Dt - D_train)

    ### parameters should be smooth
    #loss_smooth_alpha = second_derivative_loss(T, alpha)
    #loss_smooth_beta = second_derivative_loss(T, beta)
    #loss_smooth_gamma = second_derivative_loss(T, gamma)
    #loss_smooth_delta = second_derivative_loss(T, delta)

    ### update parameters
    loss_I.backward()
    loss_D.backward()
    
    #loss_smooth_alpha.backward()
    #loss_smooth_beta.backward()
    #loss_smooth_gamma.backward()
    #loss_smooth_delta.backward()
    
    optim_alpha.step()
    optim_beta.step()
    optim_gamma.step()
    optim_delta.step()

    if i % 100 == 0:
        print('=== Epoch', i, '===')
    
```
